### Group the Reason for absence 

###### discription
- <b>step1</b>: dropping the reasons for absence column to prevent multicollinearity
- <b>step2</b>: gruping the varibles or doing classificion to minimize the number of options we are analysing, so our regression can be more efficint
<ol>
  <li>1-14 are related to various diseases</li>
  <li>15-17 are related to pregnancy and giving birth</li>
  <li>18-21 poisoning</li>
   <li>22-28 light reasons for absence</li>
</ol> 
different grouping methods applies to different cases. you can chick the groups via this link """
- <b>step3</b>: adding the dummy variables to our df

### step 1

```python
#dropping the Reason for Absence column
df = df.drop(["Reason for Absence"], axis=1)
```

```python
df.head()
```

### step 2

<p>in statistics dummy variables should always sum up to 1 across each row each row have only one absence reason.</p>

now we are going to sum up all the dummy values a cross each row within the 4 groups, so that:
>- if we get 1 it means that there is a reason occurs from this specific group
>- if we get 0 that means that the reason for absence is from the other groups

```python
#grouping our dummy variables in reason_columns
#1:14 are related to vaiase diseases
#15:17 are related to pregnancy and giving birth
#18:21 poisoning
#22:28 light reasons for absance
reason_type_1 = reason_columns.loc[:, 1:14].sum(axis = 1)
reason_type_2 = reason_columns.loc[:, 15:17].sum(axis = 1)
reason_type_3 = reason_columns.loc[:, 18:21].sum(axis = 1)
reason_type_4 = reason_columns.loc[:, 22:].sum(axis = 1)
```

```python
#a sample of the result
reason_type_1
```

### step 3

```python
# adding the reason columns
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
```
