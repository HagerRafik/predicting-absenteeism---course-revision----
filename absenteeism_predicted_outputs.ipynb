## creating insert statement 

```python
df.shape
```

```python
#takes a dataframe and the table name
def insert_rows(df, table):
    insert_query = 'INSERT INTO {} VALUES '.format(table)
    for i in range(df.shape[0]):
        insert_query += '('
    
        for j in range(df.shape[1]):
            insert_query += str(df[df.columns.values[j]][i]) + ', '
    
        insert_query = insert_query[:-2] + '), '
        
    insert_query = insert_query[:-2] + ';'
    return insert_query
```

```python
insert_query = insert_rows(df, 'predicted_outputs')
insert_query
```

```python
cursor.execute(insert_query)
```
